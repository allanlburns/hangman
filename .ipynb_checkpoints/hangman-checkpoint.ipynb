```python
import time, random
from IPython.display import clear_output

def guess_word(word):
    guessed_letters = []
    required_letters = set(list(word.lower()))
    lives = 5
    
    valid_letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q',
                    'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

    while lives > 0:
        clear_output()
        print("Letters guessed:", guessed_letters)
        
        # loop to display letters:
        for char in word:
            if char == ' ':
                print(char, end="")
            elif char.lower() in guessed_letters:
                print(char, end="")
            else:
                print('-', end="")
        print('\n')

        letter = input("Which letter would you like to guess? ").lower()

        # loop to handle guessed letters
        if ' ' in required_letters:
            required_letters.remove(' ')
        if letter in word.lower() and letter not in guessed_letters:
            required_letters.remove(letter)
            guessed_letters.append(letter)
            print("Correct!")
            if len(required_letters) <= 0:
                clear_output()
                print("You win!!!!")
                print(word)
                break
            continue

        elif letter in guessed_letters:
            clear_output()
            print("You already guessed that letter. Guess another. ")
        elif letter not in valid_letters:
            print("Not a valid guess. Please enter a letter [a-z]")
            clear_output
            time.sleep(2)
        else:
            clear_output()
            print("letter not in word")
            guessed_letters.append(letter)
            lives -= 1
            print("lives remaining = ", lives)
            time.sleep(2)
            
    if lives <= 0:
        clear_output()
        print("Sorry! You lost....")
        print("The word was ", word)
        
    
        
def category_select(categories, choice):
    return random.choice(categories[choice])
    
def display_categories(categories):    
    clear_output()
    print("Categories: ")
    for category in categories.keys():
        print('\t', category)

def game():
    categories = {
        "cities": ['Boston', 'New York City', 'Tokyo', 'Berlin', 'Los Angeles', 'San Francisco', 'Chicago', 
                  'Hanoi', 'Paris', 'London', 'Fort Worth', 'New Orleans', 'Kiev', 'Reykjavik', 'Seattle', 
                  'Salt Lake City', 'Hong Kong', 'Miami', 'Mexico City', 'Rio de Janeiro', 'Portland', 'Atlanta',
                  'Nashville', 'Austin', 'St Louis', 'Philadelphia', 'Munich', 'Rome', 'Sidney', 'Moscow', 'Lagos',
                  'Cairo', 'San Diego', 'Las Vegas', 'Toronto', 'Caracas', 'Athens', 'Tunis', 'Casablanca', 'Delhi'],
        "animals": ['sugar glider','lion', 'tiger', 'bear', 'cat', 'dog', 'chupacabras', 'snake', 'zebra', 'turtle',
                   'elephant', 'puma', 'caracal', 'ocelot', 'wombat', 'squirrel', 'hippopotamus'],
         "countries": ['Mexico', 'Canada', 'Spain', 'Turkey', 'Germany', 'Ethiopia', 'Latvia', 'Switzerland',
                      'India', 'South Korea', 'Guatemala', 'Italy', 'Vietnam', 'Egypt', 'Mongolia', 'Kazakhstan',
                      'Honduras', 'Ecuador', 'Austria', 'France']
    }
    
    playing = True
    
    while playing:
        display_categories(categories)
        
        category = input("To play, please select a category. Type 'q' to quit. ").lower()
        
        if category == 'q':
            playing = False
            clear_output()
            print("Goodbye! See you soon!")
        elif category == 'animals' or category == 'countries' or category == 'cities':
            current_word = category_select(categories, category)
            guess_word(current_word)
            time.sleep(3)
        else:
            clear_output()
            print("Please enter a valid choice. ")
            time.sleep(3)
            
game()
```
